```python
# Creating an array from list
lst = [10,20,30,40]
arr1 = np.array(lst)
print(arr1, type(arr1))
print('Data type of each element in the array:- ',arr1.dtype)
print('Size(Number of elements in the array) of the array:- ',arr1.size)
print('Shape(number of RowsXColumns) of the array: ', arr1.shape)
print('Shape(number of RowsXColumns) of the array: ', arr1.ndim)
```

```python
# Array is a collection of homogeneous elements
lst = [10,20.30,30+40j,'40', [10,20]]
arr1 = np.array(lst)
print(arr1, type(arr1))
print('Data type of each element in the array:- ',arr1.dtype)
print('Size(Number of elements in the array) of the array:- ',arr1.size)
print('Shape(number of RowsXColumns) of the array: ', arr1.shape)
print('Shape(number of RowsXColumns) of the array: ', arr1.ndim)

# Int -> Float -> Complex -> String -> Object

```

```python
# Array of two dimension using list of tuple
lst_data = [('Abhishek', 101, 90909),
            ('Palani', 102, 99999),
            ('Krishna', 103, 89999)]
arr1 = np.array(lst_data)
print(arr1, type(arr1))
print(arr1.dtype)
print(arr1.ndim)
print(arr1.shape)
print(arr1.size)

```
