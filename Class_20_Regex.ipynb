```python
# match(pattern, string) - looks for the pattern inn the string in the begining and retrusn a match object.
# To extract the value from match object, we need to use group() method on match object
# if pattern s not found in the begining then it returns None.

import re
f = open('abc.txt', 'r+')
content = f.read()
pattern = r'9at'
res = re.match(pattern, content)
if res != None:
    print(res)
    print(res.group())
    print(res.span())
    print(res.start())
    print(res.end())
else:
    print('The pattern is not found in the begining of the string')


```

```python
# search(pattern, string) - looks for the pattern in the string and returns a match object of the first occurrenec of the pattern
# To extract the value from match object, we need to use group() method on match object
# if pattern s not found in the begining then it returns None.

import re
f = open('abc.txt', 'r+')
content = f.read()
pattern = r'C\w{2}'
res = re.search(pattern, content)
if res != None:
    print(res)
    print(res.group())
    print(res.span())
    print(res.start())
    print(res.end())
else:
    print('The pattern is not found in the begining of the string')
    print(res)
```

```python
# findall(pattern, string) - It finds all the occurrences of the pattern and returns a list of the all occurrences
# If the pattern is not found then it returns empty list

import re
f = open('abc.txt', 'r+')
content = f.read()
pattern = r'\wat'
res = re.findall(pattern, content)
print(res)
f.close()



```

```python
# finditer(pattern, string) - It finds all the occurrences of the pattern and returns an iterator of match object of the all occurrences
# if the pattern is not found then returns none

import re
f = open('abc.txt', 'r+')
content = f.read()
pattern = r'a[^a]{2}'
res = re.finditer(pattern, content)
for i in res:
    print(i)
    print(i.group())
    print(i.span())
    print()
f.close()



```
